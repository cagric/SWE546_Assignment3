```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pylab as plt
import pandas as pd
import sys
import heapq as hp

data = pd.read_csv(u'data/iris.txt',sep=' ')

A = np.hstack([
        np.matrix(data.sl).T, 
        np.matrix(data.sw).T, 
        np.matrix(data.pl).T, 
        np.matrix(data.pw).T])

c = np.matrix(data.c).T
xTest = np.mat([2.8, 1.4, 3.5, 1.2])

class tup:
    def __init__(self, val, idx):
        self.val = val
        self.idx = idx
    def __lt__(self, other):
        '''Redefine for max-heap'''
        return self.val > other.val
    def __le__(self, other):
        return self.val <= other.val
    def __eq__(self, other):
        return self.val == other.val
    def __ne__(self, other):
        return self.val != other.val
    def __gt__(self, other):
        return self.val > other.val
    def __ge__(self, other):
        return self.val >= other.val
    def __str__(self):
        return '{:.3},{:d}'.format(self.val,self.idx)

def evaluate(k, A, types ,iType):
    heap = []
    N = A.shape[0]
    false_positive = 0
    false_negative = 0
    true_positive = 0
    true_negative = 0
    
    for i in range(N):
        
        for l in range(k):
            hp.heappush(heap, tup(np.inf, -1))
        
        for j in range(N):
            e = A[j,:]- A[i,:]
            e = e.reshape((4,1))
            tp = tup(float(e.T*e), j)
            if tp <= heap[0]:
                hp.heapreplace(heap, tp)
        
        categories = []
        for j in range(len(heap)):
            h = hp.heappop(heap)
            categories.append(int(types[h.idx]))
        result = max(set(categories), key=categories.count)
        
        if(result == int(c[i]) and int(c[i]) == iType):
            true_positive += 1
        if(result == int(c[i]) and int(c[i]) != iType):
            true_negative += 1
        if(result != int(c[i]) and int(c[i]) == iType):
            false_negative += 1
        if(result != int(c[i]) and result == iType):
            false_positive += 1
            
    precision = true_positive / (true_positive + false_positive) * 100
    recall = true_positive / (true_positive + false_negative) * 100
    accuracy = (true_positive + true_negative) / (true_positive + true_negative + false_positive + false_negative) * 100

    print("For type:", iType)
    print("------------------")
    print("Accuracy: ", precision)
    print("Precision: ", accuracy)
    print("Recall: ", recall)
    print("\n")
    
def kNN(K, A, test, c):
    heap = []
    N = A.shape[0]
    
    for i in range(K):
        hp.heappush(heap, tup(np.inf, -1))
        
    for j in range(N):
        e = A[j,:]- test
        e = e.reshape((4,1))
        tp = tup(float(e.T*e), j)
        if tp <= heap[0]:
            hp.heapreplace(heap, tp)
    
    for m in range(K):
        h = hp.heappop(heap)
        print("Distance: ")
        print(h.val)
        print("Type")
        print(int(c[h.idx]))
        print("\n")

k = 3
#print(xTest)
kNN(k, A, xTest,c)

evaluate(k,A,c,1)
evaluate(k,A,c,2)
evaluate(k,A,c,3)
```
